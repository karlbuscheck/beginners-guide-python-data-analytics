## Merge the Data

### Setting up the survey dataset so that the code will run

```python
df_survey = pd.read_csv('cleaned_survey_super.csv', index_col=0)
```

```python
# Check to make sure the data set loaded properly
df_survey.head()
```

```python
# Check the length of the DataFrame
len(df_survey)
```

### Add the second DataFrame

```python
# Create a dataframe with two columns: 'Program' and 'Units_required'
df_programs = pd.DataFrame({'Program' : \
    ['MSIS', 'MBA', 'Master of Finance', 'Supply Chain Mgmt & Analytics', 'Master of Hacking'],\
    'Units_required' : [51, 70, 48, 49, 100]})
```

```python
# Check the DataFrame
df_programs.head()
```

```python
# Check the length of the second DataFrame
len(df_programs)
```

### Inner Merge

```python
# Merge only the rows with matching values in both DataFrames
# Note: No unmatched rows, so no NaN values created
df_survey.merge(df_programs, how='inner')
```

### Left Merge

```python
# Keep all the rows from df_survey and only the matching rows from df_programs
# Note: df_survey is the left table
# Also note: Unmatched rows from df_programs will result in NaN values
df_survey.merge(df_programs, how='left')
```

### Right Merge

```python
# Merge the two DataFrames, keeping all the rows from df_programs (the right table)
# and only matching rows from df_survey (the left table)
# Note: Unmatched rows from df_survey will result in NaN values
df_survey.merge(df_programs, how='right')
```

### Outer Merge

```python
# Merge the two DataFrames, keeping all rows from both tables, filling in NaNs
# where there is no match
# Note: Since this includes everything, it's great for combining data without losing anything
df_survey.merge(df_programs, how='outer')
```

### Merge on a shared column

```python
# Merge on a shared column, 'Program' (default is an inner merge)
# Keeps only the rows where 'Program' matches in both DataFrames
df_survey.merge(df_programs, on='Program')
```

```python
# Need to drop one duplicate and one extraneous column
df_census = df_census.drop(columns=['education', 'fnlwgt'])
```

Use Get Dummies to clean up the data set

```python
# Use pd.get_dummies() to create dummy variables for all categorical columns
df_census = pd.get_dummies(df_census, columns=['workclass', 'marital.status', 'occupation',
    'relationship', 'race', 'sex', 'native.country'], dtype=float,
    dummy_na=True)
```

```python
# Check the updated DataFrame
df_census.head()
```

### Setting the target variable

```python
# Create a copy of the DataFrame
df_census_copy = df_census.copy()
```

```python
# Write a lambda function and use the .startswith() function to turn >50K to 1.0 and <=50K to 0.0
df_census_copy['income1'] = df_census_copy.income.apply(lambda x: 1.0 if x.startswith('>') else 0.0)
```

```python
# Check the new DataFrame
df_census_copy.head()
```

```python
# Use replace function to replace'<=50K' with 0.0 and '>50K' with 1.0
df_census_copy['income2'] = df_census.income.replace({'<=50K':0.0, '>50K':1.0})
```

```python
# Drop 'income' and 'income2', the duplicate columns
df_census_copy.drop(columns=(['income', 'income2']), inplace=True)
```

```python
# Rename 'income1' to 'income'
df_census_copy.rename(columns={'income1':'income'},inplace=True)
```

```python
# Check the updated DataFrame
df_census_copy.head()
```

### Train the Decision Tree

```python
# Drop the outcome or target variable from X
X = df_census_copy.drop(columns='income')
```

```python
# Set the 'income' column as Y
Y = df_census_copy.income
```

Import the Decision Tree Classifier

```python
from sklearn import tree
```

```python
dt = tree.DecisionTreeClassifier(max_depth=2)
```

```python
# Train the decision tree classifier
dt.fit(X,Y)
```

### Visualize the tree

```python
import graphviz
# This code will visualize a decision tree dt, trained with the attributes in X and the class labels in Y
# To explain the tree:
# The tree starts by checking to see if an individual is married
# if true, it checks captial gain, if false it check education level
# then on the left branch it predicts 0.0 (don't make over 50K) if
# capital gain is less than the specified cutoff and 1.0 (do make over 50K)
# if capital gain is greater than the cutoff
# and on the right branch, the tree checks a specified education level cutoff
# if below, it predicts 0.0 and if above it predicts 1.0
dt_feature_names = list(X.columns)
dt_target_names = [str(s) for s in Y.unique()]
dot_data =tree.export_graphviz(dt, feature_names=dt_feature_names, class_names=dt_target_names, filled=True)  
graph = graphviz.Source(dot_data) 
graph
```

### Explaning the Decison Tree

The purpose of the decision tree is to predict if an individual makes more than 50K (1.0) or 50K or less (0.0). The tree makes the prediction on the basis of marital status, capital gain and education level. It first splits on marital status: if married, it checks capital gain; if not married it checks education level.

**Takeaway:** Higher capital gain and higher educational levels are strong predictors of earning more than 50K.

### Setting up the affairs dataset so that the code will run

### Clean the Data

```python
# Generate affairsBin variable, 1 if affairs>0; 0 otherwise
df_affairs['affairsBin'] = (df_affairs.affairs > 0)+ 0.0
```

```python
# Check the new DataFrame
# Note: A new column, 'affairsBin', has been added at the end
df_affairs.head()
```

### Get Dummies!

```python
# Generate dummy variables for values in 'occupation' column
# because it is a unique set of number
# Use pd.get_dummies()
# Need to update the dtype so that it's 1's and 0's not True and False
df_affairs = pd.get_dummies(df_affairs, columns=['occupation'],dtype=float)
```

```python
# Check to see if the new dummy columns have been added
df_affairs.head()
```

### Set X and Y

```python
# Drop both 'affairsBin' and 'affairs' for X
X = df_affairs.drop(['affairsBin', 'affairs'],axis=1)
```

```python
# Set affairsBin as Y
Y = df_affairs.affairsBin
```

### Split the test and training sets

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the data into 70% training set and 30% testing set
train_test_split(X,Y,test_size=30,random_state = 0)
```

```python
# Save X_train, X_test, Y_train, Y_test for future use
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cl = RandomForestClassifier(random_state = 0)

cl.fit(X_train,Y_train)
```

### Predict on the test set

```python
# Predict on the testing set (X_test)
cl.predict(X_test)
```

```python
# Display the values of the Y_test, or target variable
Y_test
```

Accuracy

```python
# Calculate accuracy: compares predicted values to actual test labels
# Returns the proportion of correct predictions
# Note: very Pythonic
(cl.predict(X_test) == Y_test).mean()
```

### Group by and sort example

```python
# This code does the following:
# Groups by 'restaurant_province;
# Calculates the average 'pizza_price' per province
# Renames the result column for carifty
# Sorts 'restaurant_province' by average 'pizza_price' in descending order
df_pizzas.groupby('restaurant_province')[['pizza_price']].mean() \
  .rename(columns={'pizza_price': 'province_avg_price'}) \
  .sort_values(by='province_avg_price', ascending=False)
```
